```python
import hashlib
import struct
import jupyter
from matplotlib import pyplot as plt
import statistics
```

```python
def gen_data(number_of_lists: int):
    res = []
    counter = 0
    for i in range(1, number_of_lists + 1):
        res.append([])
        for j in range(i):
            counter += 1
            res[i - 1].append(counter)
    return res
```

```python
def min_count(array, hash_func, k: int):
    M = [1 for _ in range(k)]
    for elem in array:
        hashed_elem = hash_func(elem)
        if hashed_elem < M[k - 1] and hashed_elem not in M:
            M[k - 1] = hashed_elem
            M.sort()
    if M[k - 1] == 1:
        return len(list(filter(lambda x: x != 1, M)))
    else:
        return (k - 1) / M[k - 1]
```

```python
def hash_1(x):
    [number] = struct.unpack("<I", hashlib.md5(x.__str__().encode()).digest()[:4])
    return number / float(0xFFFFFFFF)
```

```python
def hash_3(x):
    [number] = struct.unpack("<B", hashlib.md5(x.__str__().encode()).digest()[:1])
    return number / float(0xFF)
```

```python
def hash_2(x):
    [number] = struct.unpack("<H", hashlib.md5(x.__str__().encode()).digest()[:2])
    return number / float(0xFFFF)
```

```python
def hash_4(x):
    [number] = struct.unpack("<Q", hashlib.md5(x.__str__().encode()).digest()[:8])
    return number / float(0xFFFFFFFFFFFFFFFF)
```

```python
def ex_5b_data():
    data = gen_data(10 ** 4)
    k_set = [2, 3, 10, 100, 400]
    result = {}
    for k in k_set:
        print(k)
        data_plot = []
        for x in data:
            data_plot.append((len(x), min_count(x, hash_1, k) / len(x)))
        result[k] = data_plot
        
    return result

ex_5_data = ex_5b_data()
```

```python
def ex_5b_plot():
    for k, data in ex_5_data.items():
        print(k)
        plt.figure(k)
        plt.title(f'K={k}')
        plt.scatter([i[0] for i in data], [i[1] for i in data])
    plt.show()

ex_5b_plot()
```

```python
def ex_5c():
    data = gen_data(10 ** 4)
    k = 1
    while k < 2 ** 10:
        data_k = []
        for x in data:
            data_k.append((len(x), abs((min_count(x, hash_1, k) / len(x)) - 1)))
        result = len(list(filter(lambda y: y[1] < 0.1, data_k))) / len(data)
        if result >= 0.95:
            print(f'For K={k} Score={result}')
            return f'For K={k} Score={result}'
        if k % 10 == 0:
            print(f'For K={k} Score={result}')
        k = k + 1

ex_5c()
```

```python
def ex_6_data():
    data = gen_data(10 ** 3)

    data_res = {'1': [], '2': [], '3': []}
    for x in data:
        data_res['1'].append((len(x), abs((min_count(x, hash_1, 100) / len(x)))))
        data_res['2'].append((len(x), abs((min_count(x, hash_2, 100) / len(x)))))
        data_res['3'].append((len(x), abs((min_count(x, hash_3, 100) / len(x)))))

    return data_res

ex_6 = ex_6_data()
```

```python
def ex_6_plot():
    for i in range(1, 4):
        plt.figure(i)
        plt.scatter(*zip(*ex_6[f'{i}']))
    plt.show()
```

### HYPER LOG

```python
import math
import numpy as np
from hashlib import *


def firstOne(string):
    for c, i in zip(string, range(0, len(string))):
        if c == '1':
            return i + 1
    return 0
```

```python
def g(bin_string):
    if bin_string.find("1") < 0:
        return len(bin_string) + 1
    return bin_string.index("1") + 1


def getAlpha(m):
    if m == 16:
        return 0.673
    if m == 32:
        return 0.697
    if m == 64:
        return 0.709
    return 0.7213 / (1 + 1.079 / m)


def hyper_log_log(set, b, h):
    m = 2 ** b
    registers = np.zeros(m)
    for v in set:
        allBits = firstBits(h, v)
        j = int(allBits[0:b], 2)
        w = allBits[b:]
        registers[j] = max(registers[j], g(w))

    E = getAlpha(m) * (m ** 2) * ((np.sum(np.power(2, -registers))) ** (-1))

    if E <= (5 / 2 * m):
        V = sum(registers == 0)
        if V != 0:
            E = -m * math.log(V / m)
    if E > (2 ** 32) / 30:
        H_c = 2 ** 32
        E = - H_c * np.log((H_c - E) / H_c)

    return E


def firstBits(hash_function, string):
    string = hash_function(str(string).encode('utf-8')).hexdigest()
    return bin(int(string, 16))[-32:]
```

```python
def ex_8_data(n, b, h):
    
    data = gen_data(n)
    results = []
    for data_set in data:
        count = hyper_log_log(data_set, b, h)
        results.append((len(data_set), count / len(data_set)))
    return results

ex_8 = ex_8_data(10**4, 10, sha256)
```

```python
def ex_8_mincount_data():
    data = gen_data(10 ** 4)
    k=400
    result = []
    for x in data:
        result.append((len(x), min_count(x, hash_4, k) / len(x)))
    return result

ex_8_mincount = ex_8_mincount_data()
```

```python
def ex_8_plot():
    plt.figure(8)
    plt.scatter(*zip(*ex_8), alpha=0.5,s =1, c="k")
    plt.scatter(*zip(*(ex_8_mincount)), alpha=0.5, c="r", s=1)
    plt.show()

ex_8_plot()
```
