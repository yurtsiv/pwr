```python
import itertools
from threading import Thread

def get_procs_with_access(conf):
    have_access = []
    
    if conf[0] == conf[-1]:
        have_access.append(0)

    for i in range(1, len(conf)):
        if conf[i] != conf[i-1]:
            have_access.append(i)

    return have_access

def simulate_threads(conf, steps, index, maxes):
    n = len(conf)
    procs = get_procs_with_access(conf)
    
    if len(procs) <= 1:
        maxes[index] = steps
        return

    max_steps = [0] * len(procs)
    threads = []

    for i, proc in enumerate(procs):
        conf_copy = conf.copy()

        if proc == 0:
            conf_copy[0] = (conf[0] + 1) % (n + 1)
            threads.append(
                Thread(target=simulate, args=(conf_copy, steps + 1, i, max_steps))
            )
        else:
            conf_copy[proc] = conf[proc - 1]
            threads.append(target=simulate, args=(conf_copy, steps+1, i, max_steps)

    for t in threads:
        t.join()

    maxes[index] = max(max_steps)

def experiment(n):
    max_steps = 0
    combinations = itertools.product(*([list(range(0, n + 1))] * n))
    for i, conf in enumerate(combinations):
        if i % 10 == 0:
            print(i / ((n+1)**n) * 100, conf)

        max_steps = max(max_steps, simulate(list(conf), 0))

    print(n, max_steps)

experiment(6)
```
