```python
import math

def nakamoto(n, q):
    p = 1 - q
    la = n * (q / p)
    
    s = 0.0
    for k in range(0, n):
        s += math.exp(-la) * ((la ** k) / math.factorial(k)) * (1 - (q / p) ** (n - k))
    
    return 1 - s
```

```python
import scipy.special

def grunspan(n, q):
    p = 1 - q
    
    s = 0.0
    for k in range(0, n):
        s += (p ** n * q ** k - q ** n * p ** k) * scipy.special.comb(k + n - 1, k)
    
    return 1 - s
```

```python
import numpy as np
ns = [1, 3, 6, 12, 24, 48]
q_range = np.arange(0.01, 0.5, 0.01)
```

```python
grunspan_for_ns = {}
for n in ns:
    grunspan_for_ns[n] = []
    for q in q_range:
       grunspan_for_ns[n].append(
         (q, grunspan(n, q))
       )

nakamoto_for_ns = {}
for n in ns:
    nakamoto_for_ns[n] = []
    for q in q_range:
       nakamoto_for_ns[n].append(
         (q, nakamoto(n, q))
       )
```

```python
import matplotlib.pyplot as plt

plt.xlabel("q")
plt.ylabel("P(n)")

for n in ns:
    plt.title("n = " + str(n))
    x1, y1 = zip(*grunspan_for_ns[n])
    x2, y2 = zip(*nakamoto_for_ns[n])
 
    plt.plot(x1, y1, label="Grunspan")
    plt.plot(x2, y2, label="Nakamoto")
    plt.legend()
    plt.show()
```
