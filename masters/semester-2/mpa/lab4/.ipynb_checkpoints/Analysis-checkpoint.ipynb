```python
import numpy as np
```

```python
def read_results():
    results = {
        "fixed_points": {},
        "cycles": {},
        "records": {},
    }

    with open("./results.txt") as f:
        for section in f.read().split("n="):
            lines = section.split("\n")
            
            if not lines[0]:
                continue

            n = int(lines[0])
            results["fixed_points"][n] = []
            results["cycles"][n] = []
            results["records"][n] = []
            for line in lines[1:-1]:
                vals = line.split(",")
                results["fixed_points"][n].append(int(vals[0]))
                results["cycles"][n].append(int(vals[1]))
                results["records"][n].append(int(vals[2]))
    
    return results
```

```python
from math import log

def H(n):
    # Euler-Mascheroni constant
    gamma = 0.57721566490153286060651209008240243104215933593992
    return gamma + log(n) + 0.5/n - 1./(12*n**2) + 1./(120*n**4)
```

```python
concentration_p = 0.7

def calc_stats(values):
    mean = np.mean(values)
    var = np.var(values)
    dev = np.sqrt(var / (1 - concentration_p))

    return {
        "mean": mean,
        "var": var,
        "up_bound": mean + dev,
        "low_bound": mean - dev
    }

def analyse_results(results):
    analysis = {
        "fixed_points": {},
        "cycles": {},
        "records": {},
    }

    for n in results["fixed_points"].keys():
        analysis["fixed_points"][n] = {
            **calc_stats(results["fixed_points"][n]),
            **{ "theoretical_mean": 1 }
        }
        
        analysis["cycles"][n] = {
            **calc_stats(results["cycles"][n]),
            **{ "theoretical_mean": H(n) }
        }
        
        analysis["records"][n] = {
            **calc_stats(results["records"][n]),
            **{ "theoretical_mean": H(n) }
        }

    return analysis
```

```python
results = read_results()
```

```python
analysis = analyse_results(results)
```

```python
import matplotlib.pyplot as plt

def trend_plot(data):
    x = data.keys()
    plt.xlabel("n")
    plt.ylabel("Value")
    plt.rcParams["figure.figsize"] = (10,10)

    def plot(key, **kw):
        plt.plot(x, list(map(lambda a: a[key], data.values())), **kw)
    
    plot("mean", label="Experimental")
    plot("theoretical_mean", label="Theoretical")
    plot("up_bound", label=f"Chebyshev bounds for p = {concentration_p}", linestyle="--", color="gray")
    plot("low_bound", label=None, linestyle="--", color="gray")

    plt.legend()
    plt.grid()

def histogram(values, stats):
    plt.hist(values)
    plt.axvline(x=stats["mean"], linestyle="--", color="red", label="Mean")
    plt.axvline(x=stats["up_bound"], linestyle="--", color="gray", label=f"Chebyshev bounds for p = {concentration_p}")
    plt.axvline(x=stats["low_bound"], linestyle="--", color="gray")

    plt.legend()
```

## Fixed points

```python
trend_plot(analysis["fixed_points"])
```

```python
histogram(results["fixed_points"][500], analysis["fixed_points"][500])
```

## Cycles

```python
trend_plot(analysis["cycles"])
```

```python
histogram(results["cycles"][500], analysis["cycles"][500])
```

```python
trend_plot(analysis["records"])
```
