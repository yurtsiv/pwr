```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def parse_file(file):
    with open(file) as f:
        lines = f.read().split("\n")
        result = {}
        for line in lines:
            if line == "":
                continue

            parts = line.split(":")
            n = int(parts[0])
            values = [int(value) for value in parts[1].replace("[", "").replace("]", "").split(", ")]
            result[n] = values

        return result
```

```python
concentration = 0.9

def analyze_results(results, calc_bounds):
    analysis = {}
    
    for n in results.keys():
        values = results[n]
        mean = np.mean(values)
        up_bound, low_bound = calc_bounds(values, mean, concentration)

        analysis[n] = {
            'mean': mean,
            'up_bound': up_bound,
            'low_bound': low_bound,
            'values': values,
        }

    return analysis
```

```python
def chebyshev_bounds(values, mean, concentration):
    dev = np.sqrt(np.var(values) / (1 - concentration))
    return (mean + dev, mean - dev)
```

```python
def trend_plot(analysis, get_asympt_up):
    plt.rcParams["figure.figsize"] = (15, 10)

    ns = list(analysis.keys())
    means = [analysis[n]['mean'] for n in ns]
    up_bounds = [analysis[n]['up_bound'] for n in ns]
    low_bounds = [analysis[n]['low_bound'] for n in ns]
    asympt_up = [get_asympt_up(n) for n in ns]

    plt.xlabel("Number of balls")
    plt.ylabel("Max balls in one bin")
    plt.plot(ns, up_bounds, label=f"Bounds for p = {concentration}", linestyle="--", color="gray", linewidth='0.5')
    plt.plot(ns, low_bounds, label=None, linestyle="--", color="gray", linewidth='0.5')
    plt.plot(ns, means, label="Experimental mean")
    plt.plot(ns, asympt_up, label="Asymptotic bound", color="red")

    plt.legend()
    plt.grid()
```

```python
def task1_chernoff_bounds(values, mean, concentration):
    dev = math.e ** (-((1 - concentration) ** 2 * mean ** 2) / (2 * np.var(values)))
    return (mean + dev, mean - dev)
```

```python
task1_analysis = analyze_results(parse_file(".data/task1.txt"), task1_chernoff_bounds)
```

```python
trend_plot(
    task1_analysis,
    lambda n: 2 * math.log2(n) / (math.log2(math.log2(n)))
)
```
